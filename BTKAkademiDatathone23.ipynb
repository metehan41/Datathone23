## OutLier Handling

```python
def BoxPlotMethod(col_name, df):
    index = []
    df_variable = df[col_name]
    Q1 = df_variable.quantile(0.25)
    Q3 = df_variable.quantile(0.75)
    IQR = Q3 - Q1
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR
    low_outlier_df = df_variable[df_variable < lower_limit]
    up_outlier_df = df_variable[df_variable > upper_limit]
    print(up_outlier_df.index)
    print((low_outlier_df.shape[0] + up_outlier_df.shape[0]) / df.shape[0] )
    #df_variable[low_outlier_df] = lower_limit
    #df_variable[up_outlier_df] = upper_limit
    return up_outlier_df.index
```

```python
indx = []
for i in col_list_num:
    indx.append(BoxPlotMethod(i, df_Train))
```

## One-hot encoding and Label Encoding

```python
X = df_Train.drop(["Öbek İsmi", "index"], axis=1)
y = df_Train["Öbek İsmi"]
```

```python
X = pd.get_dummies(X)
```

```python
le = LabelEncoder()

y = le.fit_transform(y)
```

## Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_X = scaler.fit_transform(X)
```

## Train-Valid Split

```python
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(scaled_X, y, test_size=0.25, random_state=42)
```

## Feature İmportances

```python
from sklearn.ensemble import ExtraTreesClassifier
selector = ExtraTreesClassifier(random_state=42)
selector.fit(X_train, y_train)
```

```python
feature_imp = selector.feature_importances_
df=pd.DataFrame(feature_imp*100,columns=["importance"]).sort_values(["importance"], ascending=True)

importance=df["importance"].values
plt.style.use('bmh')
plt.figure(figsize=(12, 12))
sns.barplot(x = X.columns, y = importance)
plt.xlabel('INDEX NUMARASI')
plt.tick_params(labelsize=10)
plt.xticks(rotation = 90)
plt.show()
```

# Modellerin Kurulması

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier as KNN
from sklearn.metrics import mean_squared_error as mse
M = 0
dist_manhattan = []
def manhattan():
    for M in range(40):
        M = M+1
        model2 = KNN(n_neighbors = M, metric= 'manhattan')
        model2.fit(X_train, y_train)
        pred=model2.predict(X_valid)
        error = mse(y_valid,pred)
        dist_manhattan.append(error)
manhattan()
```

```python
K = 0
mse_val= []
dist_euclidean = []
def euclidean():
    for K in range(40):
        K = K+1
        model1 = KNN(n_neighbors = K, metric= 'euclidean')
        model1.fit(X_train, y_train)
        pred=model1.predict(X_valid)
        error = mse(y_valid,pred)
        dist_euclidean.append(error)
euclidean()
```

```python
plt.title('Mean Squared Error Values for kNN')
plt.plot(dist_euclidean, label = "Euclidean")
plt.plot(dist_manhattan, label = "Manhattan")
plt.xlabel('k-th neighbour')
plt.ylabel('Test Errors')
plt.legend()
```

En iyi model hyperparametreli 14 n ve manhattan distance ile elde edilmiş olup KNN modeli bu hyperparametreler ile kurulacaktır.

```python
knn = KNN(n_neighbors=14, metric="manhattan")
knn.fit(X_train, y_train)
Y_pred_knn = knn.predict(X_valid)
```
